```python
# Counting the number to severe to fatal cases for each intersection geometry, seperated by speed zones

import pandas as pd

#Cross
crossIntersectionFatal = pd.read_csv('ALLROADS_CRASH_SEVERE_CROSS_INTERSECTION.csv',encoding = 'ISO-8859-1')

#Keeping only a single record of the crash, not all other vehicles and passengers involved
crossIntersectionFatal = crossIntersectionFatal.drop_duplicates(subset=['Accident No'])

#Extracting the speed zone data from the data frame into a series and doing a count
crossSpeedZones = crossIntersectionFatal['Speed zones'].squeeze()
crossSpeedZonesCount = crossSpeedZones.value_counts()

#Removing unwanted speed zone labels
crossSpeedZonesCount = crossSpeedZonesCount.drop(labels=['Unknown', 'Camping ground'])

#Converting the index (speed zones) into integers for sorting by speed zone then printing the values
crossSpeedZonesCount.index = crossSpeedZonesCount.index.astype(str).astype(int)
crossSpeedZonesCount = crossSpeedZonesCount.sort_index(ascending=True)
print("Frequency count of severe to fatal crashes for Cross Intersection")
print(crossSpeedZonesCount)

#Same counting process performed on other intersection geometries

#Multiple
multiIntersectionFatal = pd.read_csv('ALLROADS_CRASH_SEVERE_Multiple_INTERSECTION.csv',encoding = 'ISO-8859-1')
multiIntersectionFatal = multiIntersectionFatal.drop_duplicates(subset=['Accident No'])
multiSpeedZones = multiIntersectionFatal['Speed zones'].squeeze()
multiSpeedZonesCount = multiSpeedZones.value_counts()
multiSpeedZonesCount = multiSpeedZonesCount.drop(labels=['Unknown'])
multiSpeedZonesCount.index = multiSpeedZonesCount.index.astype(str).astype(int)
multiSpeedZonesCount = multiSpeedZonesCount.sort_index(ascending=True)
print("\nFrequency count of severe to fatal crashes for Multiple Intersection")
print(multiSpeedZonesCount)

#T-juction
TIntersectionFatal = pd.read_csv('ALLROADS_CRASH_SEVERE_T_INTERSECTION.csv',encoding = 'ISO-8859-1')
TIntersectionFatal = TIntersectionFatal.drop_duplicates(subset=['Accident No'])
TSpeedZones = TIntersectionFatal['Speed zones'].squeeze()
TSpeedZonesCount = TSpeedZones.value_counts()
TSpeedZonesCount = TSpeedZonesCount.drop(labels=['Unknown', 'Camping ground'])
TSpeedZonesCount.index = TSpeedZonesCount.index.astype(str).astype(int)
TSpeedZonesCount = TSpeedZonesCount.sort_index(ascending=True)
print("\nFrequency count of severe to fatal crashes for T Junctions")
print(TSpeedZonesCount)

#Y-juction
YIntersectionFatal = pd.read_csv('ALLROADS_CRASH_SEVERE_Y_INTERSECTION.csv',encoding = 'ISO-8859-1')
YIntersectionFatal = YIntersectionFatal.drop_duplicates(subset=['Accident No'])
YSpeedZones = YIntersectionFatal['Speed zones'].squeeze()
YSpeedZonesCount = YSpeedZones.value_counts()
YSpeedZonesCount.index = YSpeedZonesCount.index.astype(str).astype(int)
YSpeedZonesCount = YSpeedZonesCount.sort_index(ascending=True)
print("\nFrequency count of severe to fatal crashes for Y Junctions")
print(YSpeedZonesCount)
```

```python
import pandas as pd
from sklearn.metrics.cluster import normalized_mutual_info_score
from sklearn.metrics.cluster import mutual_info_score

Counts = pd.read_csv('CRASH_MINOR_CROSS_INTERSECTION.csv',encoding = 'ISO-8859-1')

#Keeping only a single record of the crash, not all other vehicles and passengers involved
Counts = Counts.drop_duplicates(subset=['Accident No'])

#Removing unwanted speed zone labels and removing keeping the severity level and speed zones data
Counts.drop(Counts.loc[Counts['Speed zones']== "Unknown"].index, inplace=True)
Counts.drop(Counts.loc[Counts['Speed zones']== "Camping ground"].index, inplace=True)
Counts.drop(Counts.loc[Counts['Speed zones']== "Other speed limit"].index, inplace=True)
Counts = Counts.loc[:,['Severity level','Speed zones']]

#Creating equal width bins for the speed zones
Counts['Speed zones'] = Counts['Speed zones'].astype(int)
Counts['Speed zones'] = Counts['Speed zones'].replace([30, 40, 50], "Slow")
Counts['Speed zones'] = Counts['Speed zones'].replace([60, 70, 80], "Average")
Counts['Speed zones'] = Counts['Speed zones'].replace([90, 100, 110], "Very Fast")

print("Normalised Mutual Information Score for entire dataset with equal width binning on speed zones:")
print(normalized_mutual_info_score(Counts['Speed zones'], Counts['Severity level']))

slowSpeedZone = Counts[Counts['Speed zones'] == "Slow"]
averageSpeedZone = Counts[Counts['Speed zones'] == "Average"]
veryFastSpeedZone = Counts[Counts['Speed zones'] == "Very Fast"]

#Calculating NMI using random stratified sampling averaging method to account for different frequency distribution of bins
iterations = 100
sampleNumber = 15
aggregateNMI = 0
for i in range(iterations):
    randomState = i
    slowSample = slowSpeedZone.sample(n=sampleNumber, random_state=randomState)
    averageSample = averageSpeedZone.sample(n=sampleNumber, random_state=randomState)
    veryFastSample = veryFastSpeedZone.sample(n=sampleNumber, random_state=randomState)
    randomSample = pd.concat([slowSample, averageSample, veryFastSample])
    aggregateNMI = aggregateNMI + normalized_mutual_info_score(randomSample['Speed zones'], randomSample['Severity level'])
    
averageNMI = aggregateNMI/iterations
print("Normalised Mutual Information Score using iterative random sampling:")
print(averageNMI)
```
